Data Pre-processing

Cleaning data

```python
# reset index
dfs = dfs.reset_index(drop=True)
```

```python
dfs.head()
```

```python
# dope nulls in Gender

dfs = dfs.dropna(subset = ['Gender'])
```

```python
# dope nulls in Home Online

dfs = dfs.dropna(subset = ['Home_Ownership'])
```

```python
# dope nulls in Home Income

dfs = dfs.dropna(subset = ['Income'])
```

```python

# dope nulls in Home  Online

dfs = dfs.dropna(subset = ['Online'])
```

```python
# drop column ID
df=dfs.drop(['ID'], axis=1)
```

Train-Test Split

```python
# specify input and output attributes
X = Dropcol.drop(columns=['Age','ZIP_Code','Gender_-','Gender_F','Gender_M','Gender_O','Personal_Loan_0','Home_Ownership_Home Owner','Personal_Loan_1','Home_Ownership_Rent'], axis=1)
y = df['Personal_Loan']
```

```python
X.info()
```

```python
y.info()
```


Model Training

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python

# from matplotlib import pyplot as plt
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
model.fit(x_train, y_train)
Accuracy=model.score(x_test, y_test)*100
print("Accuracy is",Accuracy)


# cross validation - it is used for better validation of model
# eg: cv-5, train-4, test-1
score = cross_val_score(model, X, y, cv=5)
valid=np.mean(score)*100
print("Cross validation is",valid)

    # plt.plot(valid,Accuracy, label="valid,Accuracy", color='b')
```

```python
import pickle

#


# Save to file in the current working directory
pkl_filename = "Bank_classification_model.pkl"
with open(pkl_filename, 'wb') as file:
    pickle.dump(model, file)

# Load from file
with open(pkl_filename, 'rb') as file:
    pickle_model = pickle.load(file)
```

```python
y_pred = model.predict(x_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(model.score(x_test, y_test)))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
```
